```python
import pandas as pd
```

```python
dados = [['1A', '2A'], ['1B', '2B'], ['1C', '2C'],
         ['1D', '2D'], ['1E', '2E'], ['1F', '2F']]
df1 = pd.DataFrame (dados, columns=['col1', 'col2'])
df1
```

```python
dados = [['1A', '3A'], ['1B', '3B'],
         ['1C', '3C'], ['1G', '3G']]
df2 = pd.DataFrame (dados, columns=['col1', 'col3'], index=[4, 5, 6, 7])

df2
```

```python
#Unindo os DF
df1.join(df2, rsuffix='_r', lsuffix='_l')
```

```python
#Unindo os DF
df1.join(df2.set_index('col1'), on='col1', how='left')
```

```python
#Unindo os DF
df1.join(df2.set_index('col1'), on='col1', how='right')
```

```python
#Unindo os DF só o que há em comum
df1.join(df2.set_index('col1'), on='col1', how='inner')
```

```python
#Juntando os 2 DF
df1.join(df2.set_index('col1'), on='col1', how='outer')
```
