```python
import pandas as pd
import numpy as np
```

```python
dados = [[6, np.nan, 15], [np.nan, 15, np.nan], [16, 1, np.nan], [4, 19, 16],
         [np.nan, 20, 3], [6, 18, np.nan], [16, 8, 10], [8, np.nan, 3], [12, 9, 11]]
df = pd.DataFrame(dados, columns=['col1', 'col2', 'col3'])
df
```

```python
# São a mesma coisa:
# isnull
# isna
```

```python
#Filtro básico apenas valores missing
df[df['col1'].isnull()]
```

```python
#Filtro básico apenas valores missing
df[df['col1'].isna()]
```

```python
#Retornando linhas onde não tem valores missing
df[df['col1'].notnull()]
```

```python
#Retornando linhas onde não tem valores missing ~ serve para negar
df[~df['col1'].isnull()]
```

```python
#Retornando apenas Nan idependente da coluna
df[df.isnull().any(axis=1)]
```

```python
#Retornando linha com 2 valores missing
df[df.isnull().sum(axis=1) == 2]
```
