```python
import pandas as pd
import numpy as np
```

```python
dados = [[20, 14, 9], [4, 7, 8], [19, 6, 2], [12, 4, 7], [5, 11, 1],
        [15, 7, 1], [17, 7, 19], [1, 3, 13], [18, 5, 17], [6, 4, 3]]
df = pd.DataFrame(dados,
                 index=['2', '2', '4', '5', '6', '7', '8', '9', '10', '11'],
                 columns=['col1', 'col2', 'col3'])
df
```

```python
#Transformando uma coluna em índice
df.set_index('col1')
```

```python
#Criando um multi índice (multi index)
df.set_index(['col1', 'col2'])
```

```python
#Vira indice mas nao deixa de ser coluna
df.set_index('col1', drop=False)
```

```python
#Mostrando o indice inicial
df.set_index('col1',  append=True)
```

```python
df.set_index('col1', drop=False, append=True)
```

```python
#Parâmetro verify_integrity
df.set_index('col1', verify_integrity=True)
```

```python
#Resetando (reiniciando) o índice
df.reset_index()
```

```python
#Removendo a coluna index
df.reset_index(drop=True)
```

```python
#Exemplo com multi índex e multi columns

index = pd.MultiIndex.from_tuples([('classe1', 'ola'),
                                  ('classe1', 'bem-vindo'),
                                  ('classe2', 'programe'),
                                  ('classe2', 'python')],
                                 names=['classe', 'nome'])
columns = pd.MultiIndex.from_tuples([('coluna1', 'col1'),
                                    ('coluna2', 'col1')])
df1 = pd.DataFrame([(389.0, 'fly'),
                       (24.0, 'fly'),
                       (80.5, 'run'),
                       (np.nan, 'jump')],
                      index = index,
                      columns = columns)
df1
```

```python
df1.reset_index()
```

```python
df1.reset_index('classe')
```

```python
df1.reset_index('classe', col_level=1)
```

```python
df1.reset_index('classe', col_level=0)
```

```python
df1.reset_index('nome', col_level=1, col_fill='teste')
```

```python
df1.reset_index('nome', col_level=0, col_fill='teste')
```
