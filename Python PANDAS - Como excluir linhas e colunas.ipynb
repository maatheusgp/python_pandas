```python
import pandas as pd
import numpy as np
```

```python
dados = np.random.rand(9, 3)
df = pd.DataFrame(dados * 1000, columns=['col1', 'col2', 'col3'])
df
```

```python
#Removendo coluna 1
df.drop('col1', axis=1)
```

```python
#Excluindo uma linha no DF
df.drop(0, axis=0)
```

```python
#Excluindo mais de uma coluna
df.drop(['col1', 'col2'], axis=1)
```

```python
#Ignora erros que ocorreram, no caso das linhas excluidas
df.drop(['col1', 'col2'], axis=0, errors='ignore')
```

```python
#inplace a modificação é feita direta no df
df.drop('col1', axis=1, inplace=True)
```
