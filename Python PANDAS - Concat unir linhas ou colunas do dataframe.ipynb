```python
import pandas as pd
```

```python
dados = [['1A', '2A'], ['1B', '2B'], ['1C', '2C'],
         ['1D', '2D'], ['1E', '2E'], ['1F', '2F']]
df1 = pd.DataFrame (dados, columns=['col1', 'col2'])
df1
```

```python
dados = [['1A', '3A'], ['1B', '3B'],
         ['1C', '3C'], ['1G', '3G']]
df2 = pd.DataFrame (dados, columns=['col1', 'col3'])
df2
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
pd.concat([df1, df2], axis=0, ignore_index=True, join='outer')
```

```python
#Esse comando retorna só o que há de igual nas colunas
pd.concat([df1, df2], axis=0, ignore_index=True, join='inner')
```

```python
#Concat só une pelo index
pd.concat([df1, df2], axis=1)
```

```python
#Unindo colunas fora do índice
pd.concat([df1.set_index('col1'), df2.set_index('col1')], axis=1)
```

```python
#Parâmetro keys
pd.concat([df1, df2], axis=0, keys=['df1', 'df2'])
```

```python
#Parâmetro keys
pd.concat([df1, df2], axis=0, keys=['df1', 'df2'], names=[1, 2])
```
