```python
import pandas as pd
```

```python
dados = [[6, 15, 'a'], [15, 14, 'b'], [1, 14, 'c'], [19, 16, 'd'],
         [20, 3, 'f'], [18, 7, 'f'], [8, 10,'g'], [1, 3, 'h'], [9, 11, 'i']]
df = pd.DataFrame(dados, columns=['col1', 'col2', 'col3'])
df
```

```python
#Filtrando com o loc
# :(dois pontos) para mostrar que queremos filtrar todas as linhas
df.loc[:,'col1':'col3']
```

```python
#Filtrando com o iloc, com iloc passamos a posição das colunas
df.iloc[:,0:2]
```

```python
df.iloc[:,0:3]
```

```python
df.iloc[:,1:]
```

```python
#Invertendo as colunas
df.iloc[:,::-1]
```

```python
#Vendo somente a coluna 3
df.iloc[:,-1]
```
