```python
import pandas as pd
```

```python
dados = [['1A', '2A'], ['1B', '2B'], ['1C', '2C'],
         ['1D', '2D'], ['1E', '2E'], ['1F', '2F']]
df1 = pd.DataFrame (dados, columns=['col1', 'col2'])
df1
```

```python
dados = [['1A', '3A'], ['1B', '3B'],
         ['1C', '3C'], ['1G', '3G']]
df2 = pd.DataFrame (dados, columns=['col1', 'col3'], index=[4, 5, 6, 7])

df2
```

```python
#Apenas os elementos em comum
df1.merge(df2)
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, left_on='col1', right_on='col1', how='outer')
```

```python
df1.merge(df2, right_index=True, left_index=True, how='outer')
```

```python
df1.merge(df2, right_index=True, left_index=True, how='outer', suffixes=['_df1', '_df2'])
```

```python
df1.merge(df2, on='col1', how='outer', indicator=True)
```
