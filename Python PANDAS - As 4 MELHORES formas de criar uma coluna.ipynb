```python
import pandas as pd
```

```python
dados = [[1, 7], [2, 8], [3, 9], [4, 10], [5, 11], [6,12]]
df = pd.DataFrame(dados,columns=['col1', 'col2'])
df
```

```python
#Primeiro método (simples)
df['col3'] = 0
df
```

```python
df['col3'] = df['col1'] + df['col2']
df
```

```python
#Terceiro método (função insert)

df.insert(0, 'col4', 0) #posição, 'nome_coluna', valor_armazenado
df
```

```python
df.insert(2, 'col5', df['col1'], allow_duplicates=True)
df
```

```python
#Quarto método (função assign)
df.assign(col3=[13, 14, 15, 16, 17, 18], col4=df['col2'])
df
```
