```python
import pandas as pd
```

```python
dados = [[6, 15, 'a'], [15, 14, 'b'], [1, 14, 'c'], [19, 16, 'd'],
         [20, 3, 'f'], [18, 7, 'f'], [8, 10,'g'], [1, 3, 'h'], [9, 11, 'i']]
df = pd.DataFrame(dados, columns=['col1', 'col2', 'col3'])
df
```

# #Utilizando operadores do python

```python
#Somando valores
df['Total'] = df['col1'] + df['col2']
df
```

```python
#Subtraindo valores
df['Total'] = df['col2'] - df['col1']
df
```

```python
#Multiplicando valores
df['Total'] = df['col1'] * df['col2']
df
```

```python
#Dividindo valores
df['Total'] = df['col1'] / df['col2']
df
```

```python
#Exponecial valores
df['Total'] = df['col1'] ** df['col2']
df
```

```python
#Somando uma string com um n√∫mero
df['Total'] = df['col1'].astype('str') + df['col3']
df
```

```python
df['Total'] = df['col1'].astype('str') + ' ' +df['col3']
df
```
