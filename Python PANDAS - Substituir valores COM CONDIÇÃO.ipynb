```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randint(0, 20,(10, 2)), columns=['col1', 'col2'])
df
```

```python
#Função where
df.where(df > 10, 'oi')
```

```python
df.where(df < 10, 'oi')
```

```python
#Função mask substitui oq está nessa condição
df.mask(df > 12, 'maior_que_12')
```

```python
#Substituindo pelo maior valor da linha
df.where(df > 10, df.max(axis=1), axis=0)
```
