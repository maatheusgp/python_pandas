```python
import pandas as pd
```

```python
df = pd.read_excel('arquivo3.xlsx')
df
```

```python
#Retorna a média móvel
df.rolling(window=5).mean()
```

```python
df.rolling(window=5, min_periods=4).mean()
```

```python
df.rolling(window=5, min_periods=1).mean()
```

```python
df.rolling(window=5, center=True).mean()
```

```python
df.rolling(window=5, on='Data').mean()
```

```python
df.rolling(window=5, on='Views do Blog').mean()
```

```python
df['Views do Blog'].rolling(window=5).mean().head(20)
```

```python
#Passando a soma
df.rolling(window=2).sum().head(20)
```

```python
#Passando desvio padrão
df.rolling(window=2).std().head(20)
```
