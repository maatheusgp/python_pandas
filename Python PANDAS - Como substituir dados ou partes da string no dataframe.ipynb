```python
import pandas as pd
```

```python
dados = [[14, 5, 'exemplo'], [4, 11, 'teste'], [3, 12, 'pessoa'],
         [9, 8, 'teste'], [5, 19, 'exemplo'], [7, 10, 'animal'],
        [5, 6, 'pessoa'], [17, 12, 'teste'], [0, 2, 'animal'], [3, 12, 'exemplo']]
df = pd.DataFrame(dados, columns=['col1', 'col2', 'col3'])
df
```

```python
#Função replace
df.replace(5, 0)
```

```python
#Função replace
df.replace({5:0})
```

```python
#Função replace
df.replace({5:0, 12:0})
```

```python
#Função replace usando listas
df.replace([5, 12], [0, 0])
```

```python
df.replace(12, method='ffill')
```

```python
df.replace(12, method='bfill')
```

```python
df.replace(12, method='ffill', limit=1)
```

```python
#Palavra completa
df['col3'].replace('exemplo', 'oi')
```

```python
#substituindo por letra
df['col3'].replace('e', 'oi', regex=True)
```

```python
#Uma outra forma usando o str
df['col3'].str.replace('e', 'oi')
```

```python
#usando com parametro n apenas quando tem o str
df['col3'].str.replace('e', 'oi', n=1)
```
