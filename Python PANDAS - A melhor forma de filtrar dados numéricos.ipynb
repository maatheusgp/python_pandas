```python
import pandas as pd
```

```python
dados = [[6, 6, 15], [9, 15, 14], [16, 1 ,14], [4, 9, 16], [13, 20, 3],
        [6, 18, 7], [16, 8, 10], [8, 1, 3], [12, 9, 11]]
df = pd.DataFrame(dados, columns=['col1', 'col2', 'col3'])
df
```

```python
#Filtro básico condição
df[df['col1'] > 6]
```

```python
# Filtro Múltiplo
# & = e
# | = ou

df[(df['col1'] > 6) | (df['col2'] > 6)]
```

```python
df[(df['col1'] > 6) & (df['col2'] > 6)]
```

```python
#Filtrando valores entre dois valores usando o BETWEEN
df[df['col1'].between(6, 10)]
```

```python
#Filtrando valores que nao estão entre 6 e 10 usando o ~
df[~df['col1'].between(6, 10)]
```

```python
#Funções sum e max
# axis = 1 analisa as linhas
# axis=0 analisa as colunas
# retornando apenas valores que são maiores que algum valor
df[df.sum(axis=1) == 12]
```

```python
#retorna a soma das colunas sendo maior ou == a 12
df[df.sum(axis=1) >= 12]
```

```python
#Filtrando linhas cujo valor máx seja <= 12
df[df.max(axis=1) <= 12]
```

```python
#Função isin
df[df['col1'].isin([6])]
```

```python
df[df['col1'].isin([6, 12])]
```
