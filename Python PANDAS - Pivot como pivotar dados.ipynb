```python
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/st1paulama/Desktop/votacao_partido_munzona_2020_BRASIL.csv",
                 error_bad_lines=False,
                 sep=';',
                encoding='ISO-8859-1')
df                 
```

```python
a = df.groupby(['SG_PARTIDO', 'DS_CARGO', 'NR_TURNO']).sum().head(50).reset_index()
a['VOTOS'] = a['QT_VOTOS_NOMINAIS'] + a['QT_VOTOS_LEGENDA']
a = a[a['NR_TURNO'] == 1]
a.drop(['QT_VOTOS_NOMINAIS', 'QT_VOTOS_LEGENDA', 'NR_TURNO'], axis=1, inplace=True)
a
```

```python
a.pivot(index='SG_PARTIDO', columns='DS_CARGO', values='VOTOS')
```

```python
a.pivot(index='SG_PARTIDO', columns='DS_CARGO', values='VOTOS').plot(kind='bar', stacked=True)
```
